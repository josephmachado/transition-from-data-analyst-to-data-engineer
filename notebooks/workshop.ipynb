# How to Transition from Data Analyst to Data Engineer

## People Hire For Trust & Likeability 

![Hiring Criteria](./hiring-criteria.png)

* Do not spend months "just learning", start building data projects the deliver outcome to users.
* Employee's look for the following (in order of priority)
    1. **Trust & Likeability**: Can this person deliver outcomes? Would the team enjoy working with and learning from this person?
    2. **Fundamentals & Design**: Does this person know what they are doing? Will the stuff be built be a pain to maintain?
    3. **Tools/Frameworks**: Can this person quickly pick up how we build systems in our company?

* Irrespective of where you work, if you have access to data you can develop 1 & 2.

* Here's how you build them 
    1. **Trust & Likeability**: Build trust by delivering value (create datasets people use or streamline a slow/manual process).
    2. **Fundamentals & Design**: 
        - SQL, Python, Kimball Data Model, Orchestration, & Data flow
        - Design principles are built by failure analysis (another session). E.g., What happens when X fails? How to make sure the data I create is correct?                          
    3. **Tools/Frameworks**: Most tools share similiarities. Once you learn the fundamentals and design with tool A, the design is easily transferable to tool B.

## Build tables that people rely on

* Create tables that people rely on; this is your value to the business.
* In your day-to-day work, what query do you run regularly to answer an ad hoc question you’re asked or to check information yourself? Can you create a table to answer this question?
* Based on your current work, ask yourself this 
    - Who asks you questions?
    - What does the requester do with this information?
    - How often is information needed from you?

## Your first implementation should be a scheduled SQL query

* Create the entire pipeline with SQL. 
* Ideally it should be a sql query that you can schedule. 
* We assume that the data is already in the database (strive to find projects with this constraint as it will make things easier for you)


```sql
-- Example
DROP TABLE IF EXISTS summary_table;

CREATE TABLE summary_table AS
WITH fact_data AS (
    SELECT
        f.dim1_key,
        f.dim2_key,
        f.dim3_key,
        f.measure_1,
        f.measure_2,
        f.event_date
    FROM fact AS f
    WHERE f.event_date >= ADD_MONTHS(CURRENT_DATE(), -24)
),

dim1 AS (
    SELECT
        d.dim1_key,
        d.dim1_attribute_1,
        d.dim1_attribute_2
    FROM scd2_dim1 AS d
    WHERE d.is_current = TRUE
),

dim2 AS (
    SELECT
        d.dim2_key,
        d.dim2_attribute_1,
        d.dim2_attribute_2
    FROM scd2_dim2 AS d
    WHERE d.is_current = TRUE
)

SELECT
    d1.dim1_attribute_1,
    d1.dim1_attribute_2,
    d2.dim2_attribute_1,
    d2.dim2_attribute_2,
    d3.dim3_attribute_1,
    d3.dim3_attribute_2,
    SUM(fd.measure_1)              AS total_measure_1,
    SUM(fd.measure_2)              AS total_measure_2,
    COUNT(*)                       AS record_count
FROM fact_data AS fd
LEFT JOIN dim1 AS d1
    ON fd.dim1_key = d1.dim1_key
LEFT JOIN dim2 AS d2
    ON fd.dim2_key = d2.dim2_key
LEFT JOIN dim3 AS d3
    ON fd.dim3_key = d3.dim3_key
GROUP BY
    d1.dim1_attribute_1,
    d1.dim1_attribute_2,
    d2.dim2_attribute_1,
    d2.dim2_attribute_2,
    d3.dim3_attribute_1,
    d3.dim3_attribute_2;
```

* If your company has a standard scheduler (e.g. Airflow, dbt cloud, Dagster, etc), use that.

* If you do not have a scheduler use crontab if you are on Linux or Task Scheduler if you are on Windows.

* The idea here is to reduce dependency (on tools & others) and create the pipeline.


## Internal transfer to a data engineering team has a high probability of success

* Data teams want people who know the domain. 
* Interviewing is expensive, and hiring is risky. Teams only get a few hours to determine if someone is a good fit.
* Data teams can de-risk by hiring someone who has already demonstrated the ability to deliver value (hint: it’s you). * This is one of the reasons that internal transfers and referrals work much better than hiring via applications.
